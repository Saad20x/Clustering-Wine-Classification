```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
scaled_data = sc.fit_transform(data)
sc_data = pd.DataFrame(scaled_data, index=data.index, columns=data.columns)

sc_data
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(random_state=0)

visualizer = KElbowVisualizer(kmeans)

visualizer.fit(sc_data)
visualizer.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
pca_data = pca.fit_transform(sc_data)
pca_data
```

```python
pca_data = pd.DataFrame(pca_data)
pca_data
```

```python
kmeans = KMeans(random_state=0)

visualizer = KElbowVisualizer(kmeans)

visualizer.fit(pca_data)
visualizer.show()
```

```python
import matplotlib.pyplot as plt
```

```python
kmeans = KMeans(n_clusters=3, random_state=4).fit(pca_data)
sns.relplot(data=pca_data)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            c='red', edgecolor='black', s=100, alpha=0.5)
```
