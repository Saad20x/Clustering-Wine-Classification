```python
import json
import pandas as pd
import numpy as np

with open('/content/drive/MyDrive/DrivingParkingProfile-2022-01-03 (1).json')as f:
  profiles = json.load(f)


```

```python
all_trips = []
for profile in profiles:
    profile_trips = profile["trips"]
    for trip in profile_trips:
        trip["profile_ID"] = profile["id"]  # Create a new variable: you can use this to go back to the profile belonging to each trip
        all_trips.append(trip)
```

```python
profiles_df =pd.DataFrame.from_dict(profiles)
```
